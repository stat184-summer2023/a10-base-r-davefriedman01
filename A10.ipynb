```r
Counting <- data.frame(NULL)
Rows     <- c(1:3)
Cols     <- c(1:4)

for (j in 1:length(Cols)) {
  for (i in 1:length(Rows)) {
    Counting[i, j] <- i + j 
  }
}

Counting
```
